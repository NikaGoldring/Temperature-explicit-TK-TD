```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.transforms import Affine2D
```

```python
import matplotlib as mpl
mpl.rcParams['axes.spines.right'] = False
mpl.rcParams['axes.spines.top'] = False
mpl.rcParams['font.size'] = '15' 
mpl.rcParams['mathtext.default'] = 'regular'  #make sure text with subscript appears regular not italic
```

```python
IMI_LC10 = [6.79, 3.48, 1.83, 1.15, 0.85, 0.48] #µg/L
IMI_LC10low = [4.13, 2.20, 1.06, 0.60, 0.39, 0.18] #µg/L
IMI_LC10up = [10.8, 5.68, 3.09, 2.17, 1.74, 1.17] #µg/L

IMI_LC50 = [63, 32.30, 17.00, 10.70, 7.89, 4.40] #µg/L
IMI_LC50low = [41.2, 23.40, 11.70, 6.48, 4.32, 1.95] #µg/L
IMI_LC50up = [114, 50.30, 26.90, 18.50, 14.90, 9.97] #µg/L


FPF_LC10 = [8.32, 7.66, 8.15, 9.30, 10.50, 14.30] #µg/L
FPF_LC10low = [6.71, 6.04, 6.09, 6.50, 7.15, 9.30] #µg/L
FPF_LC10up = [9.99, 9.09, 9.91, 11.70, 13.40, 18.7] #µg/L

FPF_LC50 = [27.1, 19.70, 16.70, 16.20, 16.70, 19.30] #µg/L
FPF_LC50low = [21.2, 17.20, 14.10, 12.90, 12.90, 14.20] #µg/L
FPF_LC50up = [36.5, 23.60, 20.10, 20.00, 20.70, 24.5] #µg/L


XTemp = [7, 11, 15, 18, 20, 24] #degree C
```

```python
fig, ax = plt.subplots(figsize=(8,6))

# Shifting data to avoid overlapping
trans1 = Affine2D().translate(-0.2, 0.0) + ax.transData
trans2 = Affine2D().translate(-0.5, 0.0) + ax.transData
trans3 = Affine2D().translate(+0.7, 0.0) + ax.transData
trans4 = Affine2D().translate(+0.2, 0.0) + ax.transData

ax.errorbar(XTemp, IMI_LC10, yerr=[np. array(IMI_LC10) - np.array(IMI_LC10low), np.array(IMI_LC10up) - np.array(IMI_LC10)], color = 'k', markeredgecolor= 'k',  markerfacecolor = 'w', fmt='^', fillstyle = 'none',  label='IMI $LC_{10}$', markersize = 10, transform=trans2)
ax.errorbar(XTemp, IMI_LC50, yerr=[np. array(IMI_LC50) - np.array(IMI_LC50low), np.array(IMI_LC50up) - np.array(IMI_LC50)], color = 'k', fmt='^', label='IMI $LC_{50}$', markersize = 10, transform=trans1)
ax.errorbar(XTemp, FPF_LC10, yerr=[np. array(FPF_LC10) - np.array(FPF_LC10low), np.array(FPF_LC10up) - np.array(FPF_LC10)], color = 'r', markeredgecolor= 'r',  markerfacecolor = 'w', fmt='o', fillstyle = 'none',  label='FPF $LC_{10}$', markersize = 10, transform=trans4)
ax.errorbar(XTemp, FPF_LC50, yerr=[np. array(FPF_LC50) - np.array(FPF_LC50low), np.array(FPF_LC50up) - np.array(FPF_LC50)], color = 'r', fmt='o', label='FPF $LC_{50}$', markersize = 10, transform=trans3)

plt.xticks(XTemp)
plt.legend()
plt.xlabel('Temperature ['+u'\u00B0'+'C]')
plt.ylabel('$LC_{x}$'+' [µg '+u"\u00B7"+' $\mathregular{L^{-1}}$]')
plt.rc('font', size=15)          # controls default text sizes
#plt.show()

# plt.savefig("20220826_LCx-T.pdf", 
#            bbox_inches = 'tight') # so export dosn't cut of label
# plt.savefig("20220826_LCx-T.svg", bbox_inches = 'tight')
```
